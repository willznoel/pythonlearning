```python
# ref: http://www.compjour.org/warmups/govt-text-releases/intro-to-bs4-lxml-parsing-wh-press-briefings/

from bs4 import BeautifulSoup as bs
import requests

url = "https://www.gnu.ac.kr/mis/pi/prfsr/selectPrfsrIntrdView.do?mi=3186&ctgrySn=5"
response = requests.get(url, timeout = 5)

# print(response)
# print(response.content)

result_0 = bs(response.content, "html.parser")

result_1 = result_0.find_all("strong", {"class" : "name"})

print(len(result_1))

name_list = []

for t in result_1:
    # print(type(t))
    # print(t.text)
    temp = t.text.strip()
    # print(temp)
    # print(temp[:3])
    name = temp[:3]
    name_list.append(name)
```

```python
for i in name_list:
    print(i)
```

```python
import pickle

with open("data.pickle","wb") as fw:
    pickle.dump(name_list, fw)
```
